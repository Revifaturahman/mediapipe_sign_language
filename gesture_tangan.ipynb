```python
import cv2
import mediapipe as mp
import pandas as pd
from datetime import datetime

# Inisialisasi mediapipe
mp_hands = mp.solutions.hands
mp_drawing = mp.solutions.drawing_utils

# Konfigurasi penyimpanan
data = []
LABEL = "M"  # Ganti ke "sudah" atau "makan" saat merekam data lainnya

# Fungsi ambil landmark
def extract_hand_landmarks(hand_landmarks):
    coords = []
    for lm in hand_landmarks.landmark:
        coords.extend([lm.x, lm.y, lm.z])  # 21 titik * 3 = 63 nilai
    return coords
```

```python
cap = cv2.VideoCapture(0)

with mp_hands.Hands(max_num_hands=1, min_detection_confidence=0.7) as hands:
    print("Tekan 's' untuk simpan frame, 'q' untuk keluar")

    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break

        # Flip dan konversi warna
        frame = cv2.flip(frame, 1)
        rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        results = hands.process(rgb)

        # Gambar hasil
        if results.multi_hand_landmarks:
            for hand_landmarks in results.multi_hand_landmarks:
                mp_drawing.draw_landmarks(frame, hand_landmarks, mp_hands.HAND_CONNECTIONS)

        cv2.putText(frame, f"Label: {LABEL}", (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (0,255,0), 2)
        cv2.imshow("Gesture Capture", frame)

        key = cv2.waitKey(1) & 0xFF

        if key == ord('s') and results.multi_hand_landmarks:
            # Ambil 1 tangan saja
            coords = extract_hand_landmarks(results.multi_hand_landmarks[0])
            coords.append(LABEL)
            data.append(coords)
            print(f"Disimpan gesture '{LABEL}' - total data: {len(data)}")

        elif key == ord('q'):
            break

cap.release()
cv2.destroyAllWindows()
```

```python
# Buat nama kolom
columns = []
for i in range(21):
    columns += [f'x{i}', f'y{i}', f'z{i}']
columns.append('label')

# Simpan ke CSV
df = pd.DataFrame(data, columns=columns)
M = f'gesture_dataset_{LABEL}_{datetime.now().strftime("%Y%m%d%H%M%S")}.csv'
df.to_csv(M, index=False)
print(f"Disimpan ke {M}")
```
